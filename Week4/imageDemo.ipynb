```python
import openai
import os
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def generate_image(prompt, number_of_images, image_size):
    response = openai.Image.create(
      prompt= prompt,
      size=f'{image_size}x{image_size}', 
      n = number_of_images,
      response_format="b64_json"
    )
    return response
```

```python
prompt = f"""
This is a painting of a beautiful landscape. The
painting is very realistic and looks like a photograph.
The painting was painted by a famous painter named
Bob Ross. Bob Ross is a very famous painter who
has painted many beautiful paintings. He is known
for his beautiful landscapes and his paintings are
very realistic. He is a very talented painter and has
been painting for many years.
"""
number_of_images = 1
image_size = 512
```

```python
image = generate_image(prompt, number_of_images, image_size)
```

```python
print(image)
```

```python
import base64
from IPython.display import display
from PIL import Image
from io import BytesIO
```

```python
b64_string = image['data'][0]['b64_json']
```

```python
print(len(b64_string))
```

```python
image_bytes = base64.b64decode(b64_string)
```

```python
image = Image.open(BytesIO(image_bytes))
image.show()
```
