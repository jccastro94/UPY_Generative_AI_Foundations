```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return (x - 5)**2
# 25, 16, 9, 4, 1, 0
def derivative(x):
    return 2*(x - 5)
# -10, -8, -6, -4, -2, 0

```

```python
# Parameters for the gradient descent
x = 0
alpha = 0.01
precision = 0.00001
max_iters = 10000
```

```python
# Lists to store the history of x and y values 
x_history = [x]
y_history = [function(x)]
```

```python
# Gradient descent
for i in range(max_iters):
    x_old = x
    x = x - alpha * derivative(x_old)
    x_history.append(x)
    y_history.append(function(x))
    if abs(x - x_old) < precision:
        print("Converged after {} iterations. The minimum occurs at x = {}".format(i, x))
        break
```

```python
# Visualization
x_values = np.linspace(-1, 11, 1000) # 1000 points between -1 and 11
y_values = function(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Function $(x-5)^2$')
plt.scatter(x_history, y_history, color='red', marker='.', label='Gradient descent steps')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
